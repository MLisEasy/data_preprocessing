```python
import numpy as np
import pandas as pd
import os
```

```python
df=pd.read_excel('Datas.xlsx')
```

```python
df
```

```python
# the easiest way of dealing with the missing value is to drop it

df.dropna()
```

```python
# if we don't have the option of droping the missing values and need to deal with it because the data is highly important,then
#  we can fill it up using following methods

       # Replace with 0
       # Replace with mean(average)
       # Replace with Median
       # Replace with moving average
       # Replace with maximum or minimum value
```

```python
# now let's discuss one by one 
```

```python
# Replace with 0

# we can replace the missing value with simply 0

df.fillna(0)
```

```python
# as we can see the missing value is filled with 0
```

```python
# Replace with mean(average)

# we can replace the missing value with the mean

df.fillna(df.mean())
```

```python
# Replace with Median

# we can replace the missing value with median also

df.fillna(df.median())
```

```python
# Replace with moving average

# we can replace the missing value with moving average also, so here we will be taking the previous and later one values into
# consideration, you can also change it if you want to

df.fillna(df.rolling(2,min_periods=1).mean()) # here thw window is howmany values we want to take into consideration which is
# 2 here becz we want to take one from the previous and one from the later
# min_periods is min. howmany datapoints should be present to consider the operation else it will return NAN
```

```python
# Replace with maximum or minimum value

# we can replace missing value with maximum or minimum value as well

# first we will see how to replace missing value with maximum value

df.fillna(df.max())
```

```python
# now let's see how to replace with minimum

df.fillna(df.min())
```

```python
# there are few other additional methods also which are important ways of replacing the missing values

# using forward fill or backward fill
# let's check them one by one
```

```python
df.fillna(method='ffill') # here in forward fill it will replace the missing value with the previous data which is 40 here
```

```python
# now let's discuss about backward fill

df.fillna(method='bfill') # here it will replace the missing value with the value in the next row which is 80 here
```
