```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df = df[['survived','pclass','fare']]
```

```python
df.head()
```

```python
df.shape
```

```python
y = df['survived']
X = df.drop(['survived'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33)
```

```python
scaled = MinMaxScaler()
```

```python
# By default MinMaxScaler has the scaling limit of 0-1, we can change it based on our requirements
```

```python
X_train_scaled = scaled.fit_transform(X_train) # We will be training the scaler with the ttraining data and then will apply the same scaling to the testing data
```

```python
X_train_scaled
```

```python
X_test_scaled = scaled.transform(X_test)
```

```python
X_test_scaled
```
